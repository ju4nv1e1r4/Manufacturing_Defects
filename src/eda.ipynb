#### Vamos começar então a responder à algumas perguntas de negócio para que, em seguida, possamos ter um pouco mais de entendimento sobre os problemas.

##### Quais são os principais fatores que contribuem para a ocorrência de defeitos nas máquinas?

```python
correlation = df.corr()
plt.figure(figsize=(12, 9))
color = sns.color_palette('Paired')
sns.heatmap(correlation, cmap=color, linewidths=.5, annot=True, fmt='.2f')
```

#### Maiores correlações entre variáveis

Percebemos que existe correlação muito baixa entre as variáveis do dataset, mas vamos separar aquelas mais altas:

**Positiva Com Defect Status**: Production Volume(.13), Defect Rate(.25), Maintenance Hours(.30);
São correlações baixas mas ainda importantes para entendimento da situação.

**Correlações negativas**: Quality Score(-.20)
Vou guardar essa informação, apenas isolar este dado. Mas da pra ver intuitivamente que quanto maior a score de qualidade de produção, menor será a probabilidade de defeito, então já é um caminho.


##### Respondendo nossa pergunta

**Producion Volume** e **Maintenance Hours** tem correlações maiores e fazem mais sentido contribuirem para defeitos nas máquinas. Veja bem, a máquina deve suportar um volume limite de produção e, ela também deve ter um limite de manutenção, em outras palavras, mexer demais na máquina pode ocasionar defeitos.

Com isso, vou considerar criar algumas variáveis a mais: **volume de produção mensal (30 dias)** e **volume de produção semanal (7 dias)**, para assim termos a coluna **horas de manutenção por dia, semana e mês**, dai consigo chegar em duas variáveis que serão importantes, o v**olume de produção por unidade produzida** e as **horas de manutenção por unidade produzida**.

### Feature Engineering

Vamos criar as colunas abaixo:


**volume de produção mensal (30 dias)** 

**volume de produção semanal (7 dias)**

**horas de manutenção por dia**

**horas de manutenção por mês**

**volume de produção por horas de manutenção**

**horas de manutenção por unidade produzida**.


```python
df['productionvolume30dd'] = df['productionvolume'] * 30
df['productionvolume7dd'] = df['productionvolume'] * 7
df['maintenancehoursperday'] = df['maintenancehours'] / 7
df['maintenancehourspermonth'] = df['maintenancehoursperday'] * 30
df['prod_vol_permant_hour_perday'] = df['productionvolume'] / df['maintenancehoursperday']
df['mant_hour_perprod_vol_perday'] = df['maintenancehoursperday'] / df['productionvolume']
df
```

```python
correlation = df.corr()
plt.figure(figsize=(15, 12))
color = sns.color_palette('rocket')
sns.heatmap(correlation, cmap=color, linewidths=.5, annot=True, fmt='.2f')
```

##### Tomando notas das variáveis criadas e suas correlações com status de defeito

1) Existe uma correlação de .15 entre Defect Status e horas de manutenção por unidade produzida;
Embora a correlação seja positiva e moderada, isso sugere que mais horas de manutenção por unidade produzida podem estar associadas a um aumento na probabilidade de defeitos. Isso pode indicar que a manutenção excessiva ou inadequada pode ser um fator contribuinte para defeitos.

2) Existe uma correlação de -.08 entre Volume de produção diário por hora de manutenção diária.
Esta correlação negativa e baixa sugere uma relação inversa, embora fraca, entre o volume de produção diário ajustado para horas de manutenção diária. Pode indicar que um aumento na eficiência de produção (mais unidades produzidas por hora de manutenção) está ligeiramente associado a menos defeitos.

Em outras palavras, nossa EDA está indicando que deve-se dar menos manutenção, por enquanto, entretanto devo continuar a análise mais profundamente para entender outros pontos.

##### Preciso voltar para a questão da qualidade. Será que algum fator externo estaria influenciando nessa qualidade? Se sim, qual seria?
Bom, sabemos que **quanto maior a score de qualidade de produção, menor será a probabilidade de defeito**, mas a pergunta é: Quais fatores mais influenciam na qualidade?

Para responder a esta pergunta, decidir entender como outros fatores afetam a qualidade. Só pelo heatmap não deu.
Fatores:
SupplierQuality, ProductionCost, InventoryTurnover, WorkerProductivity, EnergyEfficiency, AdditiveProcessTime, AdditiveMaterialCost

```python
df['QScore_per_SupQuality'] = df['qualityscore'] / df['supplierquality']
df['QScore_per_ProdCost'] = df['qualityscore'] / df['productioncost']
df['QScore_per_InvTurnOver'] = df['qualityscore'] / df['inventoryturnover']
df['QScore_per_WorkerProd'] = df['qualityscore'] / df['workerproductivity']
df['QScore_per_EnergyEff'] = df['qualityscore'] / df['energyefficiency']
df['QScore_per_AddPrcssTime'] = df['qualityscore'] / df['additiveprocesstime']
df['QScore_per_AddMatCost'] = df['qualityscore'] / df['additivematerialcost']

df.head(7)
```

```python
correlation = df.corr()
plt.figure(figsize=(21, 18))
color = sns.color_palette('mako')
sns.heatmap(correlation, cmap=color, linewidths=.5, annot=True, fmt='.2f')
```

##### Tomando notas das variáveis de qualidade correlacionadas com fatores selecionados.

Foram criadas as variáveis: Quality Score per Supplier Quality, Quality Score per Product Cost, Quality Score per Inventory Turn over, Quality Score per Energy Efficiency, Quality Score per Additional Process Time, Quality Score per Additional Material Cost. Todas elas tem correlações negativas com Defect Status.

As correlações negativas entre as novas variáveis derivadas e o "Defect Status" reforçam a importância da "Quality Score" em diversas áreas do processo de produção e manutenção.

1) Quality Score com Quality Score per Supplier Quality: 0.91

    Uma correlação muito alta, sugerindo que a qualidade dos fornecedores tem uma grande influência no Quality Score geral.

2) Quality Score com Quality Score per Product Cost: 0.33

    Uma correlação moderada, indicando que há alguma relação entre o custo do produto e a qualidade percebida.

3) Quality Score com Quality Score per Inventory Turnover: 0.28

    Uma correlação moderada, mostrando que a rotatividade de inventário também afeta a qualidade geral.

4) Quality Score com Quality Score per Energy Efficiency: 0.30

    Uma correlação moderada, indicando que a eficiência energética está relacionada à qualidade geral.

5) Quality Score com Quality Score per Additive Process Time: 0.22

    Uma correlação baixa, sugerindo que o tempo de processamento aditivo tem uma influência menor na qualidade.

6) Quality Score com Quality Score per Additive Material Cost: 0.30

    Uma correlação moderada, mostrando que o custo do material aditivo também afeta a qualidade geral.

7) Quality Score com Quality Score per Workers Productivity: 0.91

    Uma correlação muito alta, sugerindo que a produtividade dos trabalhadores tem uma grande influência no Quality Score geral.
